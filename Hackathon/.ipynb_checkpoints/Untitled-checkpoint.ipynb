```python
# get combined data
def combined_data():
    train_csv
    test_csv
    default_status = train_csv['default_status']
    train_csv.drop(['default_status'],1,inplace=True)
    
    combined = train_csv.append(test_csv)
    combined.reset_index(inplace=True)
    
    return combined
```

```python
combined['form_field47'] = combined['form_field47'].astype('category')
combined['form_field47'] = combined['form_field47'].cat.codes
```

```python
combined.fillna(-9999, inplace=True)
```

```python
combined.isnull().sum()
```

```python
def get_data():
    global combined
    combined = combined.drop(['index', 'Applicant_ID'], 1)
    train_data = combined.iloc[:56000]
    test_data = combined.iloc[56000:]
    train = pd.read_csv('Train.csv')
    train['default_status'] = train['default_status'].astype('category')
    train['default_status'] = train['default_status'].cat.codes
    target = np.array(train['default_status'].values)
    
    return train_data, test_data, target
```

```python
import catboost as cat
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
```

```python
train_data, test_data, target = get_data()
```

```python
train_data.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data, target, test_size=0.2, random_state=42)
```

```python
# catboost
```

```python
catm = cat.CatBoostClassifier(iterations=1000, learning_rate= 0.01, eval_metric='Logloss')
```

```python
catm.fit(X_train, y_train)
```

```python
y_pred_cat = catm.predict_proba(X_test)
```

```python
y_pred_cat = y_pred_cat[:, 1]
```

```python
from sklearn.metrics import roc_auc_score, auc
```

```python
gg = roc_auc_score(y_test, y_pred_cat)
```
