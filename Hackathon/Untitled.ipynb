```python
def get_data():
    global combined
    combined = combined.drop(['index', 'Applicant_ID'], 1)
    train_data = pd.DataFrame(combined.iloc[:56000])
    test_data = pd.DataFrame(combined.iloc[56000:])
    train = pd.read_csv('Train.csv')
    train['default_status'] = train['default_status'].astype('category')
    train['default_status'] = train['default_status'].cat.codes
    target = train.default_status
    
    return train_data, test_data, target
```

```python
import catboost as cat
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
```

```python
train_data, test_data, target = get_data()
```

```python
train_data.shape, test_data.shape, target.shape
```

```python
def crossValPred(model):
    avg = 0
    test_scores = []
    for fold, (tr, te) in enumerate(skfolds.split(train_data, target), 1):
        train_x, test_x = train_data.iloc[tr], train_data.iloc[te]
        train_y, test_y = target.iloc[tr], target.iloc[te]
        
        model.fit(train_x, train_y)
        
        preds = model.predict_proba(test_x)[:, 1]
        auc = roc_auc_score(test_y, preds)
        print(f'Fold {fold}: AUC: {auc}')
        avg += auc
        
        
        test_pred = model.predict_proba(test_data)[:,1]
        test_scores.append(test_pred)
    print(f'Avg Logloss: {avg/fold}')
    return np.mean(test_scores, 0)
```

```python
skfolds= StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data, target, test_size=0.18, random_state=42)
```

# catboost

```python
catb = cat.CatBoostClassifier(iterations=4000, learning_rate= 0.01, eval_metric='AUC')
```

```python
catb.fit(X_train, y_train)
```

```python
cat_pred = crossValPred(catb)
```

# lightgbm

```python
lgbm = lgb.LGBMClassifier(random_state=42, n_estimators=2000, metric='auc', learning_rate=0.01)
```

```python
lgbm.fit(X_train, y_train)
```

```python
lgb_pred = crossValPred(lgbm)
```

```python
output = (cat_pred + lgb_pred) / 2
df_submit = pd.DataFrame()
test = pd.read_csv('Test.csv'
df_submit['Applicant_ID'] = test['Applicant_ID']
df_submit['default_status'] = output
df_submit[['Applicant_ID','default_status']].to_csv('pred_default_cat.csv', index=False)
```
